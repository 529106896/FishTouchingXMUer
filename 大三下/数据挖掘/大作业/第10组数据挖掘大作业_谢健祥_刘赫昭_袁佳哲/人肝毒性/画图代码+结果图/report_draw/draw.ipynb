```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import warnings
plt.rcParams['font.sans-serif'] = ['SimHei'] # 步骤一 替换sans-serif字体
plt.rcParams['axes.unicode_minus'] = False   # 步骤二 解决坐标轴负数的负号显示问题

warnings.filterwarnings('ignore')
```

```python
parameter_list = []
precision_0_list = []
recall_0_list = []
f1_score_list = []
precision_1_list = []
recall_1_list = []
accuracy_list = []

with open('My_vote_clf调参输出.txt', 'r') as f:
    for i in f.readlines():
        if (i.startswith('accuracy') or i.startswith('precision') or i.startswith('recall') or i.startswith('f1_score')):
            parameter_list.append(i.strip())
        elif (i.strip().startswith('0')):
            tmp = i.strip().split()
            precision_0_list.append(float(tmp[1]))
            recall_0_list.append(float(tmp[2]))
        elif (i.strip().startswith('1')):
            tmp = i.strip().split()
            precision_1_list.append(float(tmp[1]))
            recall_1_list.append(float(tmp[2]))
        elif (i.startswith('    accuracy')):
            tmp = i.strip().split()
            accuracy_list.append(float(tmp[1]))
        elif(i.startswith('   macro avg')):
            tmp = i.strip().split()
            f1_score_list.append(float(tmp[4]))

res_pd = pd.DataFrame(columns=['parameter', 'accuracy', 'f1_score', 'recall_0', 'recall_1', 'precision_0', 'precision_1'])
res_pd['parameter'] = parameter_list
res_pd['accuracy'] = accuracy_list
res_pd['precision_0'] = precision_0_list
res_pd['precision_1'] = precision_1_list
res_pd['recall_0'] = recall_0_list
res_pd['recall_1'] = recall_1_list
res_pd['f1_score'] = f1_score_list
res_pd
```

```python
res_list = res_pd['parameter'].values.tolist()
res_list
```

```python
plt.clf()
plt.figure(figsize=(15, 6))

for i in res_pd.values:
    tmp = i.tolist()
    if i[0] == 'accuracy-5' or i[0] == 'f1_score-3':
        plt.plot(i[1:], label=i[0], linewidth=2)
    else:
        plt.plot(i[1:], linestyle='--', label=i[0])

plt.legend(loc=3)
x = range(0, 6, 1)
plt.xticks(x, ('accuracy', 'f1_score', 'recall_0', 'recall_1', 'precision_0', 'precision_1'))
plt.xlabel('评价标准')
plt.ylabel('得分')
plt.title('My_vote_clf调参')
```

```python
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2.-0.12, 1.01*height, '%s' % round(float(height), 4))
```

```python
plt.clf()
plt.figure(figsize=(15, 15))
ax = plt.axes()

bar_width = 0.2
score_list = compare_pd.iloc[:, 1:].columns.tolist()

My_vote_clf_best_score = compare_pd.iloc[:1, 1:].values[0].tolist()
My_vote_select_clf_best_score = compare_pd.iloc[1:2, 1:].values[0].tolist()
My_vote_select_with_up_sample_clf_best_score = compare_pd.iloc[2:3, 1:].values[0].tolist()
baseline_score = compare_pd.iloc[3:, 1:].values[0].tolist()

autolabel(ax.bar(score_list, My_vote_clf_best_score, label='My_vote_clf', width=bar_width, align='center'))
autolabel(ax.bar(np.arange(6)+bar_width, My_vote_select_clf_best_score, label='My_vote_select_clf', width=bar_width, align='center'))
autolabel(ax.bar(np.arange(6)+bar_width*2, My_vote_select_with_up_sample_clf_best_score, label='My_vote_select_with_up_sample_clf', width=bar_width, align='center'))
autolabel(ax.bar(np.arange(6)+bar_width*3, baseline_score, label='baseline', width=bar_width, align='center'))
ax.set_title('Compare')
ax.set_xticks(score_list)
ax.set_yticks(np.arange(0, 0.95, 0.05))
ax.set_xlabel('Evaluate Parameters')
ax.set_ylabel('Evaluate score')
ax.grid(axis='y', linewidth=1)
ax.legend(loc=2, prop={'size' : 16})
plt.show()
```
